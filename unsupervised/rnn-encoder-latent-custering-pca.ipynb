```python
import numpy as np  
import pandas as pd 
import os
import matplotlib.pyplot as plt

import cv2 as cv
import librosa
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from scipy.io.wavfile import read

from tqdm.notebook import tqdm as tqdm

import torch
import torch.nn as nn

from torch.optim import Adam
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torch.utils.data import Dataset, DataLoader
import warnings

warnings.simplefilter('ignore')
```

```python
TRAIN_PART = 0.075 # part of full data used to train our model 
VAL_PART = 0.2 # part of train data for validation
```

```python
train_df = pd.read_csv('/kaggle/input/silero-audio-classifier/train.csv')
_, train_df, _, _ = train_test_split(train_df, train_df['label'].values, test_size = TRAIN_PART, stratify = train_df['label'].values)
train, val, _, _ = train_test_split(train_df, train_df['label'].values, test_size = VAL_PART, stratify = train_df['label'].values)
```

```python
def to_onehot(label, num_class = 3):
    one_hot = torch.zeros(num_class)
    one_hot[label] = 1
    return one_hot

def read_audio(path):
            sr, wav = read(path)
            assert sr == 16000
            assert len(wav) == 16000 * 3
            assert len(wav.shape) == 1
            return wav
        
def read_audio_norm(path):
            wav = read_audio(path)
            abs_max = np.abs(wav).max()
            wav = wav.astype('float32')
            if abs_max > 0:
                wav *= 1 / abs_max
            return wav

window_size = 0.02
window_stride = 0.01
sample_rate = 16000

n_fft = int(sample_rate * (window_size + 1e-8))
win_length = n_fft
hop_length = int(sample_rate * (window_stride + 1e-8))

kwargs = {
    'n_fft': n_fft,
    'hop_length': hop_length,
    'win_length': n_fft
}

def stft(wav):
    D = librosa.stft(wav,
                     **kwargs)
    mag, phase = librosa.magphase(D)    
    return mag
```

```python
class SoundDataset(Dataset):
    
    def __init__(self, df, test = False, data_path = '/kaggle/input/silero-audio-classifier/train'):
        super().__init__()
        self.data_path = data_path
        self.df = df
        self.test = test
        
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
                
        label_dict = {'speech':0,
                      'music':1,
                      'noise':2}
        
        wav = read_audio_norm(os.path.join(self.data_path, self.df.iloc[idx].wav_path))
        mag = torch.tensor(stft(wav), dtype = torch.float32).permute(1,0) #.unsqueeze(0)
        
        if not self.test:
            label = self.df.iloc[idx].target #label_dict[self.df.iloc[idx].label]
            label_one_hot = to_onehot(label)
            return mag, label_one_hot, torch.tensor(label, dtype = torch.int)
        
        return mag
```

```python
class RNNAutoEncoder(nn.Module):
    
    def __init__(self, in_size = 161, trg_len = 301, h_size = 128, n_layers = 2, bidirectional = True, device = 'cpu'):
        super().__init__()
        
        self.in_size = in_size
        self.trg_len = trg_len
        self.device = device
        self.encoder = nn.GRU(input_size = in_size, hidden_size = h_size, batch_first = True, num_layers = n_layers, bidirectional = bidirectional)
        n_directions = 2 if bidirectional else 1
        
        self.decoder = nn.GRU(input_size = 161, hidden_size = h_size, batch_first = True, num_layers = n_layers, bidirectional = bidirectional)
        
        self.decoder_output = nn.Linear(h_size*n_directions, in_size)
        
        
    def encode(self, x):
        _, h = self.encoder(x)
        return h
    
    
    def decode(self, h):
        
        batch_size = h.shape[1]
        outputs = torch.zeros(batch_size, 1, self.in_size).to(self.device)
        
        for t in range(1, self.trg_len + 1):
            output, h = self.decoder(outputs[:,-1,:].unsqueeze(1), h)
            output = self.decoder_output(output[:,-1,:]).unsqueeze(1)
            outputs = torch.cat([outputs, output], axis = 1)
            
        return outputs
    
    def forward(self, x):
        hidden = self.encode(x)
        outputs = self.decode(hidden)
        return outputs[:,1:,:]
```

```python
model_path = '../input/rnn-autoencoder/model.pth'
model = torch.load(model_path)
```

```python
train_dataset = SoundDataset(train)
val_dataset = SoundDataset(val)

train_loader = DataLoader(train_dataset, batch_size = 128, shuffle = True)
val_loader = DataLoader(val_dataset, batch_size = 128, shuffle = False)
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
```

## Create tabular datasets with latent features and save it in csv files

```python
train_latent = []
train_labels = []

tkloader = tqdm(train_loader, total = len(train_loader))

for x, _, labels in tkloader:
    x = x.to(device)
    h = model.encode(x)
    n_layers, batch_size, h_size = h.shape
    h = h.permute(1,0,2).reshape(batch_size, h_size*n_layers)
    train_latent.append(h.cpu().detach().numpy())
    train_labels.append(labels.cpu().detach().numpy())
    
X_train = np.concatenate(train_latent, axis = 0)
Y_train = np.concatenate(train_labels, axis = 0)
```

```python
sub = train_dataset.df
for i in range(X_train.shape[1]):
    sub[f"latent_{i}"] = X_train[:,i]
    
sub.to_csv('train_with_latent.csv', index = False)
```

```python
val_latent = []
val_labels = []

tkloader = tqdm(val_loader, total = len(val_loader))

for x, _, labels in tkloader:
    x = x.to(device)
    h = model.encode(x)
    n_layers, batch_size, h_size = h.shape
    h = h.permute(1,0,2).reshape(batch_size, h_size*n_layers)
    val_latent.append(h.cpu().detach().numpy())
    val_labels.append(labels.cpu().detach().numpy())
    
X_val = np.concatenate(val_latent, axis = 0)
Y_val = np.concatenate(val_labels, axis = 0)
```

```python
sub = val_dataset.df
for i in range(X_train.shape[1]):
    sub[f"latent_{i}"] = X_val[:,i]
    
sub.to_csv('val_with_latent.csv', index = False)
```

```python
for i in range(X_test.shape[1]):
    sub[f"latent_{i}"] = X_test[:,i]
    
sub.to_csv('test_with_latent.csv', index = False)
```
